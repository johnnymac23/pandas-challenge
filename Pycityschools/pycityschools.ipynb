```python
# Import dependencies
import pandas as pd

students_df = pd.read_csv('../Resources/students_complete.csv')
schools_df = pd.read_csv('../Resources/schools_complete.csv')


# Calculate total numbet of schools
total_schools = schools_df['school_name'].nunique()

# Calculate total number of students
total_students = students_df['Student ID'].nunique()

# Calculate total budget
total_budget = schools_df['budget'].sum()

# Calculate average math score
average_math_score = students_df['math_score'].mean()

# Calculate average reading score
average_reading_score = students_df['reading_score'].mean()

# Calculate percentage of students passing math (score >= 70)
passing_math_percentage = (students_df['math_score'] >= 70).mean() * 100

# Calculate percentage of students passing reading (score >= 70)
passing_reading_percentage = (students_df['reading_score'] >= 70).mean() * 100

# Calculate % oof students that are pASSING OVERALL (both math and reading score >= 70)
overall_passing_percentage = ((students_df['math_score'] >= 70) & (students_df['reading_score'] >= 70)).mean() * 100

# Create a DataFrame to summarize the district data
district_summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Passing": [overall_passing_percentage]
})

# Display the DataFrame
district_summary_df
```

```python
# Merge the school and student data into a single DataFrame
school_data_combined = pd.merge(students_df, schools_df, on="school_name")

# Group by school name and calculate key metrics
school_summary = school_data_combined.groupby('school_name').agg({
    'type': 'first',
    'size': 'first',
    'budget': 'first',
    'math_score': 'mean',
    'reading_score': 'mean'
})

# Calculate per student budget
school_summary['Per Student Budget'] = school_summary['budget'] / school_summary['size']

# Calculate % passing math
passing_math = school_data_combined[school_data_combined['math_score'] >= 70].groupby('school_name')['Student ID'].count() / school_summary['size'] * 100

# Calculate % passing reading
passing_reading = school_data_combined[school_data_combined['reading_score'] >= 70].groupby('school_name')['Student ID'].count() / school_summary['size'] * 100

# Calculate % overall passing
overall_passing = school_data_combined[(school_data_combined['math_score'] >= 70) & (school_data_combined['reading_score'] >= 70)].groupby('school_name')['Student ID'].count() / school_summary['size'] * 100

# Add passing percentages to the summary DataFrame
school_summary['% Passing Math'] = passing_math
school_summary['% Passing Reading'] = passing_reading
school_summary['% Overall Passing'] = overall_passing

# Reset index to make 'school_name' a column again
school_summary = school_summary.reset_index()

# Display the DataFrame
school_summary
```

```python
# Top Performing Schools
top_schools = school_summary.sort_values('% Overall Passing', ascending=False).head(5)
```

```python
# Worst Performing Schools
bottom_schools = school_summary.sort_values('% Overall Passing', ascending=True).head(5)
```

```python
# Display both DataFrames
top_schools, bottom_schools
```
